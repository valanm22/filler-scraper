```python
# Importing necessary libraries
import requests
from bs4 import BeautifulSoup as bs
```

```python
# Loading the webpage
r = requests.get("https://www.animefillerlist.com/shows/naruto")

# Converting to a beautiful soup object
soup = bs(r.content)

# Printing out the html
print(soup.prettify())
```
